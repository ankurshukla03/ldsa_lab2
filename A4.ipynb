```python
from pyspark import SparkContext, SparkConf

conf = SparkConf().setAppName("WordCount").setMaster("local")
sc = SparkContext(conf=conf)
```

```python
pathsrc = 'europarl-v7.bg-en.en'
rdd1 = sc.textFile(pathsrc).zipWithIndex()\
        .map(lambda x: (x[1], x[0]))
```

```python
pathtarget = 'europarl-v7.bg-en.bg'
rdd2 = sc.textFile(pathtarget).zipWithIndex()\
        .map(lambda x: (x[1], x[0]))
```

```python
#pathsrc = 'src.txt'
#pathtarget = 'target.txt'

#rdd1 = sc.textFile(pathsrc).zipWithIndex()\
#        .map(lambda x: (x[1], x[0]))
#rdd2 = sc.textFile(pathtarget).zipWithIndex()\
#        .map(lambda x: (x[1], x[0]))

rddjoin = rdd1.join(rdd2)\
            .map(lambda line:[line.split(' ') for line in line[1]])\
            .filter(lambda line:[ (line == ' ') for line in line[0]])\
            .filter(lambda line: len(line[0]) <10)\
            .filter(lambda line: len(line[0]) == len(line[1]))\
            .flatMap(lambda line: list(zip(line[0],line[1])))\
            .map(lambda word: (word,1) )\
            .reduceByKey(lambda a,b: a+b)\
            .sortBy(lambda x: -x[1])

output = rddjoin.collect()
for (word,count) in output:
    print("%s: %i" % (word,count))

#print(rddjoin.collect())
```
